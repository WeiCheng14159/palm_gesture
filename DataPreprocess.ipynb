```python
import sys  # system functions (ie. exiting the program)
import os  # operating system functions (ie. path building on Windows vs. MacOs)
import numpy as np  # matrix operations (ie. difference between two matricies)
import glob
import cv2  # (OpenCV) computer vision functions (ie. tracking)
(major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')
print('OpenCV Version: {}.{}.{}'.format(major_ver, minor_ver, subminor_ver))
```

```python
classes = {0: 'fist', 1: 'five', 2: 'point', 3: 'swing'}

kernel5x5 = np.ones((5, 5), np.uint8)
kernel3x3 = np.ones((3, 3), np.uint8)

ds_dir = os.getcwd() + '/datasets/'

for (key, gest) in classes.items():
    img_dir = ds_dir + gest
    img_path = os.path.join(img_dir, '*jpg')
    files = glob.glob(img_path)
    for f in files:
        print('read from file ', f)
        img = cv2.imread(f)
        while True:
            cv2.imshow("img", img)
            k = cv2.waitKey(-1)
            if k == 27:  #ESC
                raise SystemExit
            elif k == 91:  # press '[' to erode the image
                img = cv2.erode(img, kernel3x3, iterations=1)
            elif k == 93:  # press ']' to dilate the image
                img = cv2.dilate(img, kernel5x5, iterations=1)
            elif k == 115:  # press 's' to save the image
                done = False
                try:
                    done = cv2.imwrite(f, img)
                except:
                    pass
                if not done:
                    print('Fail to save to ', f)
                else:
                    print('Save to ', f)
                break
            elif k == 110:  # press 'n' for next image
                break
            elif k == 114:  # press 'r' to revert all changes
                img = cv2.imread(f)
            else:
                print('Key ', k, ' pressed')
```
